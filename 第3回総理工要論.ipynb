# キャリア形成 Career Development

## なぜキャリア形成は、重要なのか？　Why career developmnet is important?

　キャリア(career)とは、皆さんの生涯に渡る職業のことを指しています。大学院での教育・研究は、専攻する研究分野の専門家となるための知識と技術を体得するものです。では、専門知識を体得した先にどのような職業を選ぶかは、大きく分けて3つあります。それは、学問的な研究を継続する大学や公的研究機関、専門知識を活用し、社会に役立つ成果につなげる研究開発(R&D)を進める企業、そして、自ら研究開発から実用化までをつなげる*起業 (startup)* です。それぞれの進路選択では、考えなければならない事項があり、それに合わせたキャリア計画を練ることが必要です。

 *Career* represents an occupation undertaken for a significant period of a person's life and with opportunities for progress. The program on education and reserch at the graduate school provides you to become the professional that must have enough information and skills for specific research field. After graduating from the graduate school, you can have three options: Academia where you can continously study an academic study, industry where your expertise can create social/industrial values, and start-up where you engage yourself. Each course has its own characteristics and you have to make up the suitable career plan.

### [科学者の行動規範](http://www.scj.go.jp/ja/scj/kihan/)  Code of conduct for scientists

　[日本学術会議](http://www.scj.go.jp/ja/scj/index.html)において、**科学者とは、所属する機関に関わらず、人文・社会科学から自然科学を包含するすべての学術分野において、新たな知識を生み出す活動、あるいは科学的な知識の利活用に従事する研究者、専門職業者**と定義しています。

　社会は、18世紀の英国での[蒸気機関(Steam engine)](https://tinyurl.com/2d7can37)発明から始まる第一次[産業革命(Industrial revolution)](https://tinyurl.com/2ajn3evb)、19世紀の第二次産業革命、20世紀の第三次産業革命を経て、21世紀には[人工知能(AI)](https://tinyurl.com/4dvr447t)や[IoT (Internet of things)](https://tinyurl.com/4htm6pu7)を活用した欧州での[インダストリー4.0](https://tinyurl.com/7mcnnm7m)や、日本での[ソサエティ5.0 (Society 5.0)](https://www8.cao.go.jp/cstp/society5_0/)と大きく発展してきました。この中心にあるのが、**科学**です。**科学とは合理と実証に基づいて築かれた知識の体系**ですが、現代のように科学技術が高度化、複合化した社会においては、科学技術はシステム化されており、一般の人々は、[ブラックボックス(Black box)](https://tinyurl.com/33bb3ujj)として扱う存在となっています。しかしながら、2011年3月11日の[東日本大地震](https://tinyurl.com/4bkvw2mr)での[原子力事故 (Nuclear and radiation accidents)](https://tinyurl.com/98tvekw)や、銀行オンラインでのシステム障害などが発生すると、大きな混乱が生じ、人々は不安を感じることになります。これは、科学に関する信頼性を損ない、不信感を増大させることにつながるために、科学者は社会に対して**説明責任**を負っていると考えられます。
 
　また、科学技術の進歩は加速度的に早くなっており、その中で働いている科学者は、優れた成果を挙げるために、[捏造(fabrication)](https://tinyurl.com/enh5b63t)、[改ざん(Falsification)](https://tinyurl.com/4s5rszwz)や[盗用(pragiarism)](https://tinyurl.com/y3mnmnak)という誘惑に駆られる可能性があり、倫理規範(ethics)が重要となります。このため、日本学術会議は、2013年(平成25年)に[科学者の行動規範](http://www.scj.go.jp/ja/info/kohyo/pdf/kohyo-22-s168-1.pdf)を定めました。その重要な部分は、**科学者の責務**、**公正な研究**、**社会の中の科学**、**法令の遵守など**としてまとめられています。[こちら](http://www.scj.go.jp/ja/scj/kihan/kihan.pamflet_ja.pdf)を、目を通してください。

　**<研究活動に関する不正行為とは（[研究不正に関するガイドライン](https://www.jst.go.jp/kousei_p/outline_guidelines.html)）>**

- **捏造(fabrication)**：　存在しないデータ、研究結果等を作成すること。
- **改ざん(Falsification)**：　研究資料・機器・過程を変更する操作を行い、データ、研究活動によって得られた結果等を真正でないものに加工すること。
- **盗用(pragiarism)**：　他の研究者のアイデア、分析・解析方法、データ、研究結果、論文または用語を、当該研究者の了解もしくは適切な表示がなく流用すること。
- **その他**：　同じ研究成果の重複発表、論文著作者が適正に公開されていない不適切なオーサーシップなども不正の代表例として考えることができます。　#二重投稿は、適切な引用がされていない場合には、[*自己盗用*](https://tinyurl.com/t4jy335j)とみなされることがあります。

#### 学究 Academia

　大学等での学問的な研究を継続して進めるためには、突き詰めて考えられるかということです。**どこまでも考え続けることが出来るか**が一番重要な点です。これは、とても孤独な作業です。また、日本の大学の場合には、**研究**だけでなく、**教育**という重要な役割もあります。上のプロの研究者の要件でも述べましたが、現代の研究者は単純に研究さえしていればすむような環境ではなくなりました。学内では、数々の役割分担による雑務や学生指導があり、研究を進めていくためには、外部研究費を獲得しなければなりません。また、自分が所属している研究コミュニティへの貢献である学会活動の支援や投稿論文の査読などの作業が発生します。企業研究者と比べて、するべき業務が多岐にわたる事、時間を惜しまず作業する業務が多いことなどの欠点がありますが、自分の裁量で仕事をする事が出来るのは大きな魅力となります。
 
　When you work in academia, you have to have the confidence not to get research out of your head. The most important skill to engage in academia is to ponder on research. You have to take the roles on not only the research, but also the education.  

#### 企業 Industry

　企業とは、事業を通じて収益を得ることで存続していく[法人](https://tinyurl.com/f6m9j2fp)です。このため、企業は利益を上げることが目的であると考えられますが、現代的な定義によれば、[ピーター・ドラッガー Peter Drucker](https://en.wikipedia.org/wiki/Peter_Drucker)が述べたように、**企業とは価値を創造する組織**と言うことが出来ます。このため、[持続可能な開発目標(SDGs, Sustainable Development Goals)](https://www.mofa.go.jp/mofaj/gaiko/oda/sdgs/about/index.html)にて設定されている17の目標は企業にとっても重要な検討事項です。また、価値を生み出す行為の裏返しとして、[企業の社会的責任(CSR, Corporate Social Responsibility)](https://tinyurl.com/3fx26zwp)は、重要となってきています。
 
　企業での研究開発は、新たな価値を社会に生み出すための手段として使わていれる。このため、*純粋な基礎研究*に取り組むことは難しく、価値創造に重点をおく、すなわち応用に近い研究開発が行われています。また、通常の研究は、3年ないしは5年程度の時間の中で進めることが義務付けられています。このために、いかに優れた研究計画であっても実現に10年もかかるものは、取り組むことが難しいことになります。企業研究者に求められる資質は、**限られた研究期間の中で、社会が必要とする価値を作り出す才能**と考えられます。
 
　企業活動は、研究開発だけではありません。商品企画、マーケッティング、製品(サービス)製造、販売、営業、保守など様々な業務があります。このため、日本の会社では、皆さんの能力や適性を判断して、研究開発以外の業務に従事されることも多々あります。確かに企業でも、専門職と総合職とを分けた採用になっています。しかしながら、基本的には会社側が判断した個人ごとの資質や性向に基づいた人事を行います。これは、会社側がその人を育てることを前提としているためです。この点については、外資系企業では大きく異なります。外資系では、即戦力となる人材の配置を前提としています。このため、入社にあたっては、各人の持つ専門知識や能力が問われることになります。このため従事する業務内容(Job description)との整合性が、採用の成否の対象となります。このため応募書類には、募集されている業務内容に沿った資質や能力を記載することが成功の鍵となります。

　会社の業績が悪くなった場合には、人員削減や業務内容の取捨選択や新たな事業検討などを会社は実行します。この時には、人員削減や該当部署の廃止あるいは新規部署の設置などが行われ、これに伴い日本の場合には、配置転換が行われることで、出来るだけ人材を減らさないような配慮がなされます。これに対して、米国では、個々人の持つ資質と能力を判断して、解雇(layoff)や採用(employment)が行われます。日本企業の賃金体系は成果主義(performace evaluation)に移行しつつありますが、採用は、いまだに必ずしも能力主義(ability-based evaluation)ではありません。

## 産学官連携とは？　What is academia, industry, and government collaboration?

　大学での研究に関する資金は、[文部科学省](https://www.mext.go.jp)の[科学研究費(科研費)](https://www.jsps.go.jp/j-grantsinaid/)、文部科学省以外の省庁も含めて応募が可能となる[競争的研究資金](https://www8.cao.go.jp/cstp/compefund/)、そして企業や自治体、国の研究機関と共同で進める共同研究があります。競争的資金は、日本の科学戦略を決める[総合科学技術会議・イノベーション会議](https://www8.cao.go.jp/cstp/gaiyo/index.html)(以前は、総合科学技術会議と呼ばれていた)において、2003年(平成15年)の[提言](https://www8.cao.go.jp/cstp/output/iken030421_1.pdf)に従って作られた研究資金制度で、一般に科学研究費よりも大きな研究費となるものです。令和元年(2019年)の科学研究費は、[2万9000件を採択し2154億円を配分](https://www.mext.go.jp/a_menu/shinkou/hojyo/1422129.htm)したが、競争的研究資金は、それよりも大幅に少ない採択件数において、当初予算で、[約4430億円](https://www8.cao.go.jp/cstp/compefund/kyoukin_r1-2.pdf)を配算しています。

　国からの研究資金は膨大な額ではありますが、[科学技術・学術政策研究所(NISTEP)](https://www.nistep.go.jp)の[科学技術指標2018](https://www.nistep.go.jp/sti_indicator/2018/RM274_11.html)によりますと、企業を含めた我が国全体の研究開発費総額に占める政府の研究資金は20%を切り、近年は諸外国に比べて減少の傾向にあります。
 
　一方、企業から見た場合、研究支出資金に対して、その研究成果は小さくなりつつあります。この理由は、イノベーションを興すためには基礎研究が重要ですが、基礎研究は必ずしも大きな成果を得られるとは限らないためです。また、日本企業の経済成長が諸外国に比べて低いために、益々、基礎研究を減らさなければならないというジレンマに陥っています。更に、欧米諸国では、[オープンイノベーション(Open Innovation)](https://www.nedo.go.jp/content/100879995.pdf)を活用することで、短時間で大きな変革を進めています。
 
 　このようなことから、これまで企業独自で進めてきた基礎研究を大学との共同研究に切り替えることで、研究コストを減らし、大学の知を活用して効率的に研究を進めることが増加してきています。実際、諸外国の場合には、企業が大学に対して大きな研究費を出して成果を上げてきています。
  
  また、国立の研究機関も競争的研究資金により、以前よりも潤沢な研究費を獲得していますが、研究員を増やすことは難しい現実があります。このため、大学との連携を強化して、学生を受け入れることで研究を推進する取り組みが広がっています。このように、企業と大学、公的研究機関と大学、あるいは三者での研究協力を総称して[産学官連携](https://www.mext.go.jp/a_menu/shinkou/sangaku/sangakua.htm)と呼んでいます。
  
> 【備考】産官学という呼び方もありますが、連携とつなげる場合には、産学官連携が一般的です。

# 産学官連携共同研究 Collaborative research 

　共同研究(collaborative research or joint research)は、皆さんの研究にも大きく関わってくるものです。英語では、collective researchと、joint researchがありますが、少し意味合いが違います。例えば、競争的研究資金で実行される研究は、それぞれの機関が研究分担を行い、一つの大きな研究を実施するため、Joint researchの意味合いが強くなります。これに対して、連携研究は、異なる研究機関が、お互いに対等な関係(equal partnership)で、相互に関心を持ち実益のある研究(mutual interesting and benefitable research)を進めるものです。ここでは、後の方に相当する産学官連携共同研究について説明を致します。


## 共同研究の円滑な推進の検討事項　Issues to discuss for proceeding a collaborative research 

　では、共同研究を始めるために何が必要でしょうか。研究を進めるためには、必ず[リソース（resources)](https://biz.trans-suite.jp/27121)が必要です。リソースとは、研究を進めるために必要となるもの全てを指します。研究課題を理解するためには、**専門的な知識**が必要であり、また研究を推進するための*アイデア*を出したり、実際に研究を行う**研究者**は必須です。また、必要な機材や材料を購入するための**研究資金**も必要となります。更に、忘れてはならないのは、**研究期間**です。いつまでに研究成果を上げるか、研究を終了するのかをきちんと決めておかなければなりません。

　リソースを決定すれば、共同研究は円滑に進められるでしょうか？研究を進めるためには、**研究計画(research plan)** は必須です。研究計画には、研究課題名と研究に携わる研究者の名前とその責任者である研究代表者(PI, principle investigator)の決定と共に、研究課題を明らかにするための具体的な研究項目等が盛り込まれます。研究計画に従って研究を進めたとしても、研究には失敗がつきものであり、また予定通りに研究が進む保証もありません。このために研究計画を変更したり、研究予定を変えることも必要となるかもしれません。そのような変更や修正を**判断**するのが[研究代表者(PI)の役目](https://tinyurl.com/su83mz7f)となります。また、研究代表者は、より良い研究成果を上げられるように研究チームをまとめていくことも重要な役割の一つです。

　共同研究を進める中で、新たな知見や発見があり、研究に関係した技術や技能などの獲得やそれらを体系化した知識の蓄積が出来上がります。このような共同研究から出てきた成果を合わせて、**研究成果**と呼び、その具体的な内容を**成果物(deliverable(s))** と言います。共同研究では、成果物の適正な管理も重要な事項です。
 
　以上のように、共同研究において光量すべき多数の事項があり、それぞれに**責任と義務**があります。これらが確実に遂行されるように交わす取り決めのことを**共同研究契約書**と言います。文科省が挙げている具体的な[契約書例](https://www.mext.go.jp/b_menu/hakusho/nc/t20020329006/img/t20020329006_y0000002.pdf)をご覧ください。九州大学の基本的な共同研究契約文書と締結を進めるための手順については、[学術研究・産学官連携本部 AirImaQ](https://airimaq.kyushu-u.ac.jp/ja/airimaq/greet.php)の[受託研究・共同研究](https://airimaq.kyushu-u.ac.jp/ja/sky/jk_flow.php)をご覧ください。受託研究とは、民間企業等から、本学の多様な研究者や研究施設設備に着目していただき、本学が委託を受けて行う研究であり、本学の研究者は職務として当該研究を行います。
 
 　共同研究契約書は、研究課題ごとに考慮すべき事項が異なることも多くあり、相手機関との記載内容のすり合わせについて専門の事務職員が指示を行い交渉していきます。[経済産業省](https://www.meti.go.jp/index.html)に掲載されている[事例](https://www.meti.go.jp/press/2020/06/20200630006/20200630006-10.pdf)を挙げておきます。なお、文部科学省と経済産業省は、大学における産学官連携を推進するために、[**産学官連携による共同研究強化のためのガイドライン**](https://www.meti.go.jp/press/2020/06/20200630004/20200630008.html)をまとめています（2020年6月に改訂）。このガイドラインについての推進のために、2021年1月にシンポジウムが開催されており、[関連資料](https://www.meti.go.jp/policy/innovation_corp/symposium_seminar.html)も出ています。

### **知的財産権 (Intellectual property right, IPR)**

 知的財産権とは、特許権、実用新案権、意匠権及び上記各権利の登録を受ける権利、著作権(著作権法第27条及び第28条に規定される権利を含む)、ノウハウの使用権、成果有体物の所有権その他研究成果に関する一切の権利を指しています。[特許庁](https://www.jpo.go.jp/index.html)のWebにて、[特許権、実用新案件、意匠権](https://www.jpo.go.jp/system/basic/index.html)の定義と活用方法について説明されています。特許権の範囲が及ぶのは特許権を登録した国の中だけです。また、その権利は、国内では基本的に20年間だけ保証されるます。しかしながら、医療応用や諸外国での法律制度との整合性のために、特許権の存続期間の延長も5年まで可能となっています。詳細は、参考2をご覧ください。共同研究における[著作権 copyright](https://tinyurl.com/4j9p69mb)は、論文や書籍、また研究のために作られたプログラムやアルゴリズムなどが該当します。
 
　[*知的財産権*](https://tinyurl.com/zdfwh2zw)とは、複製が可能である[無体物](https://tinyurl.com/4mpp4uk3)を保護するための権利であり、特に、特許権、実用新案権、意匠権、商標権を総称して、[産業財産権(工業所有権)](https://tinyurl.com/c282ku5m)と言います。産業財産権は、登録される前におおやけになってしまうと権利を失うという性質があります。例えば、新しい発明に対して特許権を得るためには、**研究会での発表や論文掲載の前に、特許登録することが必須**です。

　【参考】[クリエイティブ・コモンズ](https://tinyurl.com/33jp5rr8)　米国、ハーバード大学の[ローレンス・レッシグ(Lawrence Lessig)](https://lessig.org)教授が提唱した*著作物の適切な再利用*を促す仕組みです。法律や技術に関する専門的な知識がなくても、自分の希望する条件を組み合わせることで、自分の作品をインターネットを通じて世界に発信することができる画期的なライセンスシステムとなっています。概略はこちらを[ご覧](https://creativecommons.jp/faq/#a2)下さい。
 
  【参考2】[特許権の存続期間の延長](https://www.jpo.go.jp/system/laws/rule/guideline/patent/tukujitu_kijun/kaitei2/document/encho_shitsumon_1903/ix_nashi.pdf) 2020年3月10日以降に出願された特許からは、出願から５年後または審査請求から３年後のいずれか遅い日（基準日）を超えた後に特許登録されたときは、その基準日よりオーバーした日数分から一部控除した日数だけ、特許権の存続期間を出願日から２０年という限界を超えて延長することができますと変更されました。これは、[TPP (環太平洋パートナーシップ協定](https://www.mofa.go.jp/mofaj/gaiko/tpp/index.html)の締結により発生した事です。詳細は、[特許権の存続期間の延長について（蒲田特許事務所)](https://www.kamada-pat.com/blog/?p=168)を御覧下さい。
 

### **秘守義務 (confidentiality)**

　知的財産権以外でも、共同研究においては他者や他機関に知られては損失が生じる情報を**秘密情報**と呼びます。このような秘密情報を守ることを秘守義務と呼びます。共同研究契約書の中には、必ず秘守義務が入っています。秘守義務のみを守るための契約を[秘密保持契約(Non-disclosure agreement, NDA)](https://tinyurl.com/4fvunz6h)と呼びます。秘密情報を関する法律である[不正競争防止法](https://www.meti.go.jp/policy/economy/chizai/chiteki/index.html)では、[**営業秘密**](https://www.meti.go.jp/policy/economy/chizai/chiteki/trade-secret.html)と呼びます。秘守義務を破った場合には、損害賠償や罰則も生じますので、気をつけて下さい。
 
　【実例】秘密保持義務(NDA)は、米国の大学では最重要なこととされています。米国の大学において共同研究を実施した研究者が、共同研究を終えて帰国しました。この研究者は、米国において実施した共同研究を更に進めた研究を日本において実施して、それに関する特許の申請を行いました。研究者と所属する研究機関は問題がないという認識でしたが、米国の大学から数十億円に上る損害賠償を請求されました。その理由は、NDAの中にあった研究資料を持ち帰ってはいけないという規約に反して、研究者は実験ノートを持ち帰ったためでした。

# アントレプレナーシップ教育 Entrepreneurship education

　アントレプレナーシップ(entreprenurship)をご存知ですか？学部一年生から博士課程の大学院生まで履修出来る、国内の大学の中で一番充実したカリキュラムを九州大学は提供しています。起業家のことをアントレプレナーと言いますが、アントレプレナーシップ教育は、起業することを目的とした教育ではありません。*不確実な現代社会の中で生き抜く術*と考えています。以下に、詳しく説明いたします。

## アントレプレナーシップ教育とは？ What is entreprenureship education?

　アントレプレナーシップは、よく*起業家活動*と訳されますが、実際には広い解釈を持つ言葉です。[ベンチャーキャピトル(Venture capital)](https://tinyurl.com/2f539s8w)からの支援を得たスタートアップ(start-up)企業と考える人もいれば、[小規模企業(SMEs)](https://tinyurl.com/3kuxh4by)活動と考える人もいます。また、企業内起業（コーポレート・アントレプレナーシップ [Corporate entrepreneurship](https://sloanreview.mit.edu/article/the-four-models-of-corporate-entrepreneurship/)）のことと考える方もいるでしょう。
 
　ハーバード・ビジネス・スクールの[ハワード・スティーブンソン(Howard Stevenson)](https://en.wikipedia.org/wiki/Howard_H._Stevenson)教授は、「[アントレプレナーシップとは、コントロール可能な資源を超越して機会を追求すること](https://www.dhbr.net/articles/-/1947) [(entrepreneurship is the pursuit of opportunity beyond resources controlled.)](https://hbr.org/2013/01/what-is-entrepreneurship)と定義しています。詳しくは、クリックして[Harverd Business Review](https://www.dhbr.net)を読んてみてください。もっと簡単に表現した場合には、[「0から1を生み出す(zero to one)」](https://business.nikkei.com/atcl/opinion/16/041400010/110700014/)ことと言われています。

 **自分の考えを深めることで、新たなアイデアを創出し、それを社会や科学・産業に役立つ価値として創造すること**ということが出来ます。これは、会社を興し事業を展開することだけではなく、まさに**研究活動**にも当てはまります。従って、アントレプレナー教育は、普遍的な考え方や技能を培うものであるといえます。
 
 [九州大学のアントレプレナーシップ教育](https://qrec.kyushu-u.ac.jp/abouts/)では、
 
- 自ら機会を発見・創造して、目標と道筋を構想する能力ある人材

- 個人としての自立意識を持つ人材
- 社会や世界を幅広く俯瞰できる知識と能力を持つ人材（T字型人材、グローバル人材）
- 知識を社会で活かす意欲と具体的価値を創造する能力ある人材（MOT人材）

という実物像の育成を目指しています。

## 研究を効率的に進めるためには？ <br>How to proceed your research project effectively</br>

研究を効率的に進めるためには、各種のアプリケーションソフトウェアを活用することは大切です。アプリケーションソフトウェアには、有料のものと無料のものがあります。有料のものが良いように感じるかもしれませんが、[オープンソースソフトウェア (Open source software, OSS)](https://tinyurl.com/85bj5ch6)は、多くのボランティアに支えられた活動で成り立ち信頼性もあり有益です。通常無料で使うことが可能です。


### 文献管理ツール　Management tools for references

- [Mendeley](https://www.elsevier.com/ja-jp/solutions/mendeley) ![Mendeley](figs/mendeley.jpg) Mendeley（メンデレー）は、学術論文の管理とオンラインでの情報共有を目的とした無料の文献管理ツールです。Windows、Mac、Linuxに対応するデスクトップ版と、オンラインでどこからでも利用できるウェブ版を組み合わせて使用できます。iOS（iPad、iPhone）にも対応しています。九州大学附属図書館の[ここ](https://www.lib.kyushu-u.ac.jp/ja/databases/mendeley)に説明があります。[エルゼビア](https://www.elsevier.com/ja-jp)と大学との契約がありますので、100GBまでの容量を利用することが可能です。また、[Scopus](https://www.elsevier.com/ja-jp/solutions/scopus)との連携も可能となっています。
- [Zotero](https://www.zotero.org) ![Zotero](figs/Zotero.png) Zoteroとはオープンソースソフトウェアで作られた文献管理用ソフトです。文献の管理とWordでの引用文献リストの作成が可能です。データは、文献のPDFファイルを含めて300MBまでクラウドで無料で保存が可能で、デバイス間で共有出来ます。PDFファイルを含めると300MBはすぐに容量限度に達します。それ以上の容量が必要な場合には有料のサービスもあります。皆さんがお持ちのMicrosoft社の[OneDrive](https://www.microsoft.com/ja-jp/microsoft-365/onedrive/onedrive-for-business)や[Dropbox](https://www.dropbox.com/)と組み合わせて、大量の文献PDFファイルも合わせて無料で使うことも可能です。Zoteroは文献情報やタグに基づいた非常に柔軟な検索機能を有しています。PDFファイルの本文から検索することも可能です。また、検索条件を保存しておくことで、動的に更新される文献コレクションとして利用することもできます。これを使ったことで、いちいち著者の名前、出版年、を記録し、引用スタイルに従ってポチポチ打っていく作業が皆無になったのです。

### 論文執筆ツール Academic drafting tools

　皆さんの研究成果は、最終的には[論文](https://tinyurl.com/mpdp58av)として発表しなければなりません。論文は、グラフや表、そして挿絵などと本文とが[組版](https://tinyurl.com/xbpbhbf)により整えられて掲載されています。このように実際に掲載される状態で論文を見ることが出来る論文執筆ツールとして、[$\LaTeX$](https://www.latex-project.org)があります。雑誌ごとに定められたテンプレートを使うことで、印刷された状態を確認しながら論文作成が可能となります。[マイクロソフト社](https://www.microsoft.com/ja-jp)の[Word](https://www.microsoft.com/ja-jp/microsoft-365/word)用のテンプレートもありますが、実際に印刷と同じ形態と品質での組版は、$\LaTeX$が一番です。

- [$\LaTeX$](https://tinyurl.com/5xrkkuks) Macintoshをお使いの場合には、[MacTeX](http://tug.org/mactex/)を、Windowsをお使いの場合には、[TeX Live](https://texwiki.texjp.org/?Microsoft%20Windows)をお使いになると良いと思います。$\LaTeX$の情報は、インターネットでも得ることが可能ですが、[三重大学の奥村晴彦](https://oku.edu.mie-u.ac.jp/~okumura/)先生が書かれた[LaTeX2ε美文書作成入門](https://www.amazon.co.jp/gp/product/4297117126)が解説書としては有名です。$\LaTeX$は、ボランティアのグループにより常に最新版に改良されており、LaTeX2ε美文書作成入門もアップデートされて最新版（現在は第8版）になっていっています。なお、印刷に間に合わない詳しい情報は、[こちら](https://texwiki.texjp.org)で情報を得ることが出来ます。


> 【一口メモ】$\LaTeX$は、マイクロソフトのWordに比べて、記号や数式の作成には入力が難しく大変です。しかしながら、組版された現行の美しさは、惚れ惚れとするものです。マイクロソフトWordの文章は、[プレインテキスト](https://tinyurl.com/bfmhyp6)ではありません。紙と異なり電子データは永久に残すことが出来ると思われますが、[プレインテキスト](https://tinyurl.com/bfmhyp6)でないと、時代が変わり、ソフトウェアが陳腐化すると使えなくなります。$\LaTeX$の場合には、文字に記号を付与するだけですので、永遠に残すことが可能です。資料の永続性の観点からも$\LaTeX$を勧めることが出来ます。
